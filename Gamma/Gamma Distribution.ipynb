### Definition

The **Gamma Distribution** is continuous and is the sum of exponentials, where you determine the probability it will take a certain amount of time to see a fixed number of events occur. This distribution takes two parameters, the number of events and the rate, $\upsilon$, at which the events occur. We would say 

<center>$W_r \sim Gamma(r, \upsilon)$</center>

Alternatively, you could also exchange the rate with the mean, $\theta$, such that $\theta = \frac{1}{\upsilon}$. In this case,

<center>$W_r \sim Gamma(r, \theta)$</center>


### Probability Density Function

The probability of waiting $w$ time to see $r$ events occur is:

$f(w)$ = $\frac{\upsilon^r w^{r-1}}{\Gamma (r)}e^{-\upsilon w}$


This pdf can be adapted so that the parameter is the mean:

$f(w)$ = $\frac{w^{r-1}}{\theta^r \Gamma (r)}e^{-w/\theta}$

For $r \in \mathbb{Z}^+, \Gamma (r) = (r-1)!$

### Descriptive Statistics 

The Gamma distribution is simply the sum of independent exponentials ($W_r = Y_1 + Y_2 + ... + Y_r$ where $Y_i$'s are independent). Therefore, the mean of the gamma distribution is:


$E[W_r] = E[\sum_{i=1}^r Y_i] = \sum_{i=1}^r \frac{1}{\upsilon} = \frac{r}{\upsilon} = r\theta$

And the variance of the gamma distribution is:

$Var[W_r] = Var[\sum_{i=1}^r Y_i] = \sum_{i=1}^r \frac{1}{\upsilon^2} = \frac{r}{\upsilon^2} = r\theta^2$

```python
# import packages
import numpy as np
import matplotlib as mpl
from scipy.stats import gamma
from matplotlib import pyplot as plt

# set figure defaults
mpl.rcParams['figure.dpi'] = 150
plt.rcParams['figure.figsize'] = (12.0/2, 8.0/2)

# establish parameters
k_values = [1, 2, 2, 3, 5]
theta_values = [2, 1, 2, 1, 0.5]
linestyles = ['-', '--', ':', '-.']
x = np.linspace(1E-6, 10, 1000)

# plot the distributions
fig, ax = plt.subplots(figsize=(5, 3.75))

for k, t in zip(k_values, theta_values):
    dist = gamma(k, 0, t)
    plt.plot(x, dist.pdf(x), 
             label=r'$k=%.1f,\ \theta=%.1f$' % (k, t))

plt.xlim(0, 10)
plt.ylim(0, 0.5)

plt.xlabel('$x$')
plt.ylabel(r'$p(x|k,\theta)$')
plt.title('Gamma Distribution')

plt.legend(loc=0)
plt.show()
```
